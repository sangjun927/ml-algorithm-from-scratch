# 1 Binary classification with PyTorch

## 1)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from torch.utils.data import TensorDataset
from torch.utils.data import Dataset, DataLoader
import torch
```

```python
df = pd.read_csv("data_hw2/default-credit-card-clients.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df_encoded = pd.get_dummies(df, columns=["EDUCATION", "MARRIAGE"])
```

```python
df_encoded
```

## 2)

```python
np.random.seed(3)
msk = np.random.rand(len(df_encoded)) < 0.8
train = df_encoded[msk].reset_index()
valid = df_encoded[~msk].reset_index()
```

```python
train.info()
```

```python
valid.info()
```

## 3)

```python
X_train = train.drop('default_payment', axis=1)
y_train = train['default_payment']
X_valid = valid.drop('default_payment', axis=1)
y_valid = valid['default_payment']

# Initialize the StandardScaler
scaler = StandardScaler()

X_train_normalized = scaler.fit_transform(X_train)
X_valid_normalized = scaler.transform(X_valid)
```

## 4)

```python
# Convert the NumPy arrays to PyTorch tensors
X_train_tensor = torch.tensor(X_train_normalized.astype(np.float32))
y_train_tensor = torch.tensor(y_train.values.astype(np.float32))
X_valid_tensor = torch.tensor(X_valid_normalized.astype(np.float32))
y_valid_tensor = torch.tensor(y_valid.values.astype(np.float32))

# Use TensorDataset to create dataset objects for PyTorch
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
valid_dataset = TensorDataset(X_valid_tensor, y_valid_tensor)
```

## 5)

```python
# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=1000, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=1000)
```

## 6)

```python
from torch import nn

class LogisticRegression(nn.Module):
    def __init__(self, num_features):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(num_features, 1)
    
    def forward(self, x):
        return torch.sigmoid(self.linear(x))

# Initialize the model
num_features = X_train_normalized.shape[1]
model = LogisticRegression(num_features)

# Test the model on a batch of data
x, y = next(iter(train_loader))
with torch.no_grad():
    y_pred = model(x)
    print(y_pred[:5])  # Display predictions for the first 5 examples
```

## 7)

```python
from sklearn.metrics import roc_auc_score, log_loss

def evaluate_model(model, data_loader):
    model.eval()  # Set the model to evaluation mode
    all_predictions = []
    all_targets = []
    
    with torch.no_grad():
        for inputs, targets in data_loader:
            outputs = model(inputs)
            all_predictions.extend(outputs.view(-1).tolist())
            all_targets.extend(targets.view(-1).tolist())
    
    auc_roc = roc_auc_score(all_targets, all_predictions)
    logloss = log_loss(all_targets, all_predictions)
    return logloss, auc_roc
```

## 8)

```python
def train_loop(model, train_dl, valid_dl, optimizer, epochs):
    for epoch in range(epochs):
        model.train()  # Set the model to training mode
        total_loss = 0
        
        for inputs, targets in train_dl:
            optimizer.zero_grad()  # Reset the gradients
            outputs = model(inputs)  # Forward pass
            loss = nn.BCELoss()(outputs.view(-1), targets)  # Compute the loss
            loss.backward()  # Backward pass
            optimizer.step()  # Update model parameters
            
            total_loss += loss.item() * inputs.size(0)
        
        train_loss = total_loss / len(train_dl.dataset)
        
        # Evaluate on validation set
        valid_loss, valid_auc_roc = evaluate_model(model, valid_dl)
        
        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, "
              f"Validation Loss: {valid_loss:.4f}, Validation AUC-ROC: {valid_auc_roc:.4f}")
```

## 9)

```python
from torch.optim import Adam

# Initialize the logistic regression model
model = LogisticRegression(num_features)

# Initialize the optimizer
optimizer = Adam(model.parameters(), lr=0.001)  # Adjust learning rate as needed

# Specify the number of epochs
epochs = 10

# Train the model
train_loop(model, train_loader, valid_loader, optimizer, epochs)
```

## 10)

```python
class TwoLayerModel(nn.Module):
    def __init__(self, num_features, hidden_size):
        super(TwoLayerModel, self).__init__()
        self.linear1 = nn.Linear(num_features, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, 1)
    
    def forward(self, x):
        x = self.relu(self.linear1(x))
        x = torch.sigmoid(self.linear2(x))
        return x

# Initialize the two-layer model
hidden_size = 10  
model_two_layers = TwoLayerModel(num_features, hidden_size)

# Train the two-layer model
optimizer_two_layers = Adam(model_two_layers.parameters(), lr=0.001)
train_loop(model_two_layers, train_loader, valid_loader, optimizer_two_layers, epochs)
```
